# Environment Setup

```python
exp_type = "anime"
detailed = exp_type!="anime2"
import pickle
with open(f'{exp_type}_exp.pkl', 'rb') as f:
    experiment = pickle.load(f)
vitucb1 = experiment.agents[2]
vitucb1.name="Partial"
    
vitucb2 = ViT_UCB_Diag(experiment.env, lora_r=20, alpha=85, lambda_reg=4, lora_alpha=24, detailed=detailed)
vitucb2.name="Diag"
print(vitucb1.numel, vitucb2.numel)
experiment.agents = [vitucb1, vitucb2]
experiment.results = Path("diag results")
experiment.results.mkdir(exist_ok=True)
```
